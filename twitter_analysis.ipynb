## Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## Loading the dataset

```python
df = pd.read_csv("twitter_training.csv", header = None)
df.head()
```

```python
df.columns = ['id', 'product', 'label', 'comment']
df.head()
```

## Analysis

```python
plt.figure(figsize=(20, 8))
sns.countplot(df, hue='label', x = 'product', palette= 'rocket')
plt.xticks(rotation=90)
```

Insight: Customers have given negative comments mostly among all type of comments. However some products like Borderlands, CallOfDutyBlackopsColdWar, Overwatch, Xbox(Xseries), PlayStation5(PS5), CS-GO, AssassinsCreed, Hearthstone, RedDeadRedemption(RDR) and Cyberpunk2077 has maximum no of positive comments and most of these are games or related to games. It can be concluded that most of the social media commenters are passionate about games.

Products like NBA2K, Verizon, FIFA, MaddenNFL, CallOfDuty, TomClancysRainbowSix, HomeDepot have huge amount of negative comments which have even passed the number of all other type of comments.
